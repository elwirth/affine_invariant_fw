```python
import matplotlib.pyplot as plt
C = 2
D = 3
gap = []






def eta(k):
    return 2 / (k + 2)

def recursion(n_iters: int = 1000000, q: float = 2.0, r: float = 1):
    gap = [C]
    for i in range(0, n_iters):
        gap.append((1-eta(i)) * gap[-1] + D * eta(i)**q * gap[-1]**r)


    fig = plt.figure()
    fig.set_figheight(8)
    fig.set_figwidth(8)

    plt.rcParams['font.family'] = 'DeJavu Serif'
    plt.rcParams['font.serif'] = ['Computer Modern Roman']
    plt.rcParams['text.usetex'] = True

    plt.plot(gap)
    plt.grid(True)

    plt.xscale("log")
    plt.yscale("log")
    plt.show()
```

```python
recursion(q = 1.5)
```
